```python
import os
import wget
import shutil
import pandas as pd
# import jax; jax.config()
```

# データのダウンロードと整形

```python
date = "20230323"
url = f"https://www.jpx.co.jp/markets/derivatives/option-price/data/ose{date}tp.zip"
fname_csv = f"data/ose{date}tp.csv"
if os.path.isfile(fname_csv):
    print("already downloaded")
else:
    os.makedirs("data", exist_ok=True)
    fname_zip = f"data/ose{date}tp.zip"
    wget.download(url, out=fname_zip) # download
    shutil.unpack_archive(fname_zip, extract_dir="data") # unzip
```

```python
# csv header
# rearranged from https://www.jpx.co.jp/markets/derivatives/option-price/tvdivq00000014eu-att/head.csv
heads = ["商品コード", "商品タイプ", "限月", "権利行使価格", "予備", "プットオプション_銘柄コード", "プットオプション_終値", "プットオプション_予備", "プットオプション_理論価格", "プットオプション_ボラティリティ", "コールオプション_銘柄コード", "コールオプション_終値", "コールオプション_予備", "コールオプション_理論価格", "コールオプション_ボラティリティ", "原資産終値", "基準ボラティリティ"]
# dataframe
df = pd.read_csv(fname_csv, names=heads)
df = df[df["商品コード"].str.contains("NK225E")]; df.head()
```
